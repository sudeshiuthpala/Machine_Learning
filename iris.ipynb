```python
#k-nearest neighbour Algorithm
#strings thiyagena labelswalata yanne na
#0-setosa 1-verginika 2-Versicolor(values of labels-target)
#features - sepal.length,sepal.width,petal.lenght,petal.width(values of features-data)
import pandas as pd
dataset=pd.read_csv(r"iris.csv").values
print(dataset.shape)#(150,5)
print(dataset)
```

```python
data=dataset[:,0:4]
target=dataset[:,4]
print(data.shape)
print(target.shape)
print(data)
print(target)
```

```python
#prediction walata danna issella select karaganna ona samples (related).randomly selected
#not all data
#train from samples data and target
#after that test for ithuru data walata with target hrida kiyala. predicted answer compare karanawa 
#actual targets
# it passe predicted ewata balanawa-final
#library sklearn

from sklearn.model_selection import train_test_split

train_data,test_data,train_target,test_target=train_test_split(data,target,test_size=0.2)
#train 150*20%,test 150*80% 
#data split to train,test data/target split to tarin,test target
```

```python
#select algorithm
from sklearn.neighbors import KNeighborsClassifier
#from sklearn.svm import SVC => Support Vector
model=KNeighborsClassifier()
```

```python
#train
model.fit(train_data,train_target)
```

```python
#test
predicted_target=model.predict(test_data)#getting predictions from the algorithm for test data
```

```python
#evaluate
print('Actual target',test_target)
print('predicted target',predicted_target)#compare
```

```python
#get accuracy with predicted and actual
from sklearn.metrics import accuracy_score
acc=accuracy_score(test_target,predicted_target)
print(acc)
#give 96% accuracy it is best amount
```
